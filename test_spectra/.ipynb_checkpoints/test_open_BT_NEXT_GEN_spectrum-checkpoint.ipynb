```python
def block_reduce_cube(cube, binning_factor, func=np.sum):
    '''
    Resample a cube by doing spatial binning.
    binning_factor should be greater than 1. 
    binning_factor set to 2 gives a spatial binning of 2x2 pixels^2.
    '''
    #print(cube)
    CUBE_BINNED = []
    for i in range(len(cube)):
        CUBE_BINNED.append(block_reduce(cube[i], binning_factor, func=func))
    return np.array(CUBE_BINNED)
```

```python
def binning_spectral_bad(cube, spectral_binning_factor=2, idx_mid_bad=18):
    cube = cube[1:-1]
    idx_mid_bad -=1
    n = np.shape(cube)[0]
    nb_frames_to_be_removed = n%spectral_binning_factor
    
    idx_mid_bad_beg = idx_mid_bad - nb_frames_to_be_removed//2 # first frame to not consider
    idx_mid_bad_end = idx_mid_bad + nb_frames_to_be_removed//2 # last frame to not consider 
    
    if nb_frames_to_be_removed % 2 == 1:
        idx_mid_bad_beg = idx_mid_bad-1
        
    print('The spectral binning factor is {}.\nThe number of frames is {}.'.format(spectral_binning_factor, n))
    print('We consider the frames 2 to {} ({} channels) and {} to 38 ({} channels), count starting to 1.'.format(1+idx_mid_bad_beg,
                            idx_mid_bad_beg, 2+idx_mid_bad_end, 37 - idx_mid_bad_end))
    
    cube1 = cube[:idx_mid_bad_beg]
    cube2 = cube[idx_mid_bad_end:]
    
    print('Shape of cube 1:', np.shape(cube1))
    print('Shape of cube 2:', np.shape(cube2))
    return
    
    
```

```python
def binning_spectral(cube, wgrid, spectral_binning_factor=2, remove_first_last_frame=1, display=1):
    '''
    Apply spectral binning. Sum "spectral_binning_factor" frames to get one. 
    Default value, spectral_binning_factor=2.
    Remove the first and last frame if "remove_first_last_frame" is set to 1.
    Display information if "display" set to 1.
    One has to provide the "cube" and the wavelength grid ("wgrid").
    '''
    if display: print('== Apply spectral binning ==')
        
    if remove_first_last_frame:
        if display: print('Remove the first and last frame of the IFS cube.')
        frames_add = 2
        cube = cube[1:-1]
    else: frames_add = 0
        
    n = np.shape(cube)[0]
    nb_frames_to_be_removed = n%spectral_binning_factor
    
    idx_beg = nb_frames_to_be_removed//2 # first frame to consider
    idx_end = n - nb_frames_to_be_removed//2 # last frame to consider 
    
    if nb_frames_to_be_removed % 2 == 1:
        idx_beg += 1
        
    if display: print('The spectral binning factor is {}.'.format(spectral_binning_factor, n))
    #print('Number of frames to remove {}/{}, index beg={} and index end={}.'.format(nb_frames_to_be_removed+2, n+2, idx_beg, idx_end))
    if display: print('Number of frames to remove {}/{}.'.format(nb_frames_to_be_removed+frames_add, n+frames_add))
    
    cube = cube[idx_beg:idx_end]
    m = np.shape(cube)[0] # should be equal to "idx_end - idx_beg"
    CUBE_BINNED, W_BINNED = [], []
    if display: print('We consider the frames {} to {} ({} channels), count starting to 1.'.format(frames_add//2+1+idx_beg,
                            frames_add//2+idx_end, m))
    
    for i in range(0,m,spectral_binning_factor):
        CUBE_BINNED.append( np.sum(cube[i:i+spectral_binning_factor],axis=0) )
        W_BINNED.append( np.mean(wgrid[i:i+spectral_binning_factor]) )
    
    return np.array(CUBE_BINNED), np.array(W_BINNED)
    
```
