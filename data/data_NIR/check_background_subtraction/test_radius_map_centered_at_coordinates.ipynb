# Create radius map with the center located at coordinates (xm,ym)

```python
def compute_im_rad_grid_coord(im, xm=0, ym=0):
    '''
    Compute a 2D radius grid with the same shape than the input image 'im'.
    Optional parameter 'even_or_odd' only considered as 'even'. (!)
    '''
    nx = np.shape(im)[1]
    ny = np.shape(im)[0]
    xend = nx-xm
    yend = ny-ym 
    xbeg = -xm
    ybeg = -ym
    x,y = np.arange(xbeg,xend,1), np.arange(ybeg, yend,1)
    xs,ys = np.meshgrid(x, y, sparse=True)
    zs = np.sqrt(xs**2 + ys**2)
    return zs
```

```python
mm = np.zeros((40,40))
im_rad = compute_im_rad_grid_coord(mm,10,20)
plt.imshow(im_rad, interpolation='nearest', origin='lower')
plt.show()
```
